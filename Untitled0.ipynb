```python
import os
import cv2
import numpy as np
from tqdm import tqdm

# Define input and output paths
# Assuming 'data_path' is defined in a previous cell, if not, define it here:
# data_path = '/content/drive/MyDrive/new/rice@/data'
input_path = data_path
output_path = os.path.join(data_path, 'preprocessed_images')
os.makedirs(output_path, exist_ok=True)

# ==============================
# Background Removal (Simple Green Masking)
# ==============================
def remove_background(image):
    hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

    # Green color range
    lower_green = np.array([25, 40, 40])
    upper_green = np.array([90, 255, 255])

    mask = cv2.inRange(hsv, lower_green, upper_green)
    result = cv2.bitwise_and(image, image, mask=mask)

    return result

# ==============================
# Image Enhancement (CLAHE)
# ==============================
def enhance_image(image):
    lab = cv2.cvtColor(image, cv2.COLOR_BGR2LAB)
    l, a, b = cv2.split(lab)

    clahe = cv2.createCLAHE(clipLimit=3.0, tileGridSize=(8,8))
    cl = clahe.apply(l)

    merged = cv2.merge((cl,a,b))
    enhanced = cv2.cvtColor(merged, cv2.COLOR_LAB2BGR)

    return enhanced

# ==============================
# Noise Removal (Gaussian Blur)
# ==============================
def remove_noise(image):
    return cv2.GaussianBlur(image, (5,5), 0)

# ==============================
# Augmentation
# ==============================
def augment_image(image):
    rows, cols, _ = image.shape

    # Rotation
    M_rotate = cv2.getRotationMatrix2D((cols/2, rows/2), 20, 1)
    rotated = cv2.warpAffine(image, M_rotate, (cols, rows))

    # Scaling
    scaled = cv2.resize(image, None, fx=1.2, fy=1.2)

    # Translation
    M_translate = np.float32([[1,0,20],[0,1,20]])
    translated = cv2.warpAffine(image, M_translate, (cols, rows))

    # Zoom (Crop center)
    h, w = image.shape[:2]
    zoom = image[int(h*0.1):int(h*0.9), int(w*0.1):int(w*0.9)]
    zoom = cv2.resize(zoom, (w, h))

    return [rotated, scaled, translated, zoom]

# ==============================
# Normalize (0-1 range)
# ==============================
def normalize_image(image):
    return image / 255.0

# ==============================
# Main Loop
# ==============================
for class_folder in os.listdir(input_path):
    class_path = os.path.join(input_path, class_folder)

    if not os.path.isdir(class_path):
        continue

    save_class_path = os.path.join(output_path, class_folder)
    os.makedirs(save_class_path, exist_ok=True)

    for img_name in tqdm(os.listdir(class_path)):
        img_path = os.path.join(class_path, img_name)
        image = cv2.imread(img_path)

        if image is None:
            continue

        # 1. Background Removal
        image = remove_background(image)

        # 2. Enhancement
        image = enhance_image(image)

        # 3. Noise Removal
        image = remove_noise(image)

        # 4. Resize (224x224)
        image = cv2.resize(image, (224,224))

        # 5. Normalize
        norm_image = normalize_image(image)

        # Save original processed image
        save_path = os.path.join(save_class_path, img_name)
        cv2.imwrite(save_path, image)

        # 6. Augmentation
        aug_images = augment_image(image)

        for i, aug in enumerate(aug_images):
            aug = cv2.resize(aug, (224,224))
            aug_name = img_name.split('.')[0] + f"_aug{i}.jpg"
            cv2.imwrite(os.path.join(save_class_path, aug_name), aug)

print("✅ Preprocessing Completed Successfully!")
```

```python
import os
import shutil
import random

# ==============================
# Dataset Path
# ==============================
# Corrected data_path to point to the directory containing preprocessed images
data_path = "/content/drive/MyDrive/new/rice@/data/preprocessed_images"
output_base = "/content/drive/MyDrive/new/rice@/final_split"

train_path = os.path.join(output_base, "train")
test_path = os.path.join(output_base, "test")
val_path = os.path.join(output_base, "val")

# Create folders
for path in [train_path, test_path, val_path]:
    os.makedirs(path, exist_ok=True)

# ==============================
# Split Dataset
# ==============================
for class_name in os.listdir(data_path):
    class_folder = os.path.join(data_path, class_name)

    if not os.path.isdir(class_folder):
        continue

    images = os.listdir(class_folder)
    random.shuffle(images)

    total = len(images)
    train_split = int(0.70 * total)
    test_split = int(0.15 * total)

    train_imgs = images[:train_split]
    test_imgs = images[train_split:train_split + test_split]
    val_imgs = images[train_split + test_split:]

    # Create class folders
    for split_path in [train_path, test_path, val_path]:
        os.makedirs(os.path.join(split_path, class_name), exist_ok=True)

    # Copy images
    for img in train_imgs:
        shutil.copy(os.path.join(class_folder, img),
                    os.path.join(train_path, class_name, img))

    for img in test_imgs:
        shutil.copy(os.path.join(class_folder, img),
                    os.path.join(test_path, class_name, img))

    for img in val_imgs:
        shutil.copy(os.path.join(class_folder, img),
                    os.path.join(val_path, class_name, img))

print("✅ Dataset Split Completed!")
```

```python
def count_images(folder):
    total = 0
    for class_name in os.listdir(folder):
        class_folder = os.path.join(folder, class_name)
        if os.path.isdir(class_folder):
            total += len(os.listdir(class_folder))
    return total

train_count = count_images(train_path)
test_count = count_images(test_path)
val_count = count_images(val_path)

print("📊 Dataset Distribution:")
print("Training Images  :", train_count)
print("Testing Images   :", test_count)
print("Validation Images:", val_count)
print("Total Images     :", train_count + test_count + val_count)
```

```python
import tensorflow as tf
from tensorflow.keras.applications import VGG19
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.optimizers import Adam

# ==============================
# Dataset Path
# ==============================
train_dir = "/content/drive/MyDrive/new/rice@/final_split/train"
val_dir   = "/content/drive/MyDrive/new/rice@/final_split/val"
test_dir  = "/content/drive/MyDrive/new/rice@/final_split/test"

IMG_SIZE = 224
BATCH_SIZE = 64

# ==============================
# Data Generator
# ==============================
train_datagen = ImageDataGenerator(rescale=1./255)
val_datagen   = ImageDataGenerator(rescale=1./255)
test_datagen  = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='categorical'
)

val_generator = val_datagen.flow_from_directory(
    val_dir,
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='categorical'
)

test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=False
)

num_classes = train_generator.num_classes

# ==============================
# Load VGG19 Base Model
# ==============================
base_model = VGG19(weights='imagenet',
                   include_top=False,
                   input_shape=(224, 224, 3))

# Freeze Base Model
for layer in base_model.layers:
    layer.trainable = False

# ==============================
# Add Custom Layers
# ==============================
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(512, activation='relu')(x)
x = Dropout(0.5)(x)
output = Dense(num_classes, activation='softmax')(x)

model = Model(inputs=base_model.input, outputs=output)

# ==============================
# Compile Model
# ==============================
model.compile(
    optimizer=Adam(learning_rate=0.001),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

model.summary()

# ==============================
# Train Model
# ==============================
history = model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=10
)

# ==============================
# Evaluate on Test Data
# ==============================
test_loss, test_acc = model.evaluate(test_generator)
print("Test Accuracy:", test_acc)
```

```python
import tensorflow as tf
from tensorflow.keras.applications import VGG19
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint

# ==============================
# Dataset Path
# ==============================
train_dir = "/content/drive/MyDrive/new/rice@/final_split/train"
val_dir   = "/content/drive/MyDrive/new/rice@/final_split/val"
test_dir  = "/content/drive/MyDrive/new/rice@/final_split/test"

IMG_SIZE = 224
BATCH_SIZE = 64

# ==============================
# Data Generator
# ==============================
train_datagen = ImageDataGenerator(rescale=1./255)
val_datagen   = ImageDataGenerator(rescale=1./255)
test_datagen  = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='categorical'
)

val_generator = val_datagen.flow_from_directory(
    val_dir,
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='categorical'
)

test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=False
)

num_classes = train_generator.num_classes

# ==============================
# Load Base Model
# ==============================
base_model = VGG19(weights='imagenet',
                   include_top=False,
                   input_shape=(224, 224, 3))

# Freeze first
for layer in base_model.layers:
    layer.trainable = False

# ==============================
# Add Custom Layers
# ==============================
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(512, activation='relu')(x)
x = Dropout(0.5)(x)
output = Dense(num_classes, activation='softmax')(x)

model = Model(inputs=base_model.input, outputs=output)

# ==============================
# Compile (Initial Training)
# ==============================
model.compile(
    optimizer=Adam(learning_rate=0.001),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

print("🚀 Initial Training (Frozen Base Model)")
history1 = model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=5
)

# ==============================
# Fine Tuning
# ==============================

# Unfreeze last 6 layers
for layer in base_model.layers[-6:]:
    layer.trainable = True

# Recompile with lower learning rate
model.compile(
    optimizer=Adam(learning_rate=0.00001),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

print("🔥 Fine-Tuning Started")

history2 = model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=10
)

# ==============================
# Evaluate
# ==============================
test_loss, test_acc = model.evaluate(test_generator)
print("🧪 Final Test Accuracy:", test_acc)
```
